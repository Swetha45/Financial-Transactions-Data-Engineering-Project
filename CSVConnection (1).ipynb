## Import the Libraraies

```python
from pyspark.sql.functions import*
```

## Mounting the data

```python

spark.conf.set("fs.azure.account.auth.type.adlfdevbankdomain.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.adlfdevbankdomain.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.adlfdevbankdomain.dfs.core.windows.net", "client.id")
spark.conf.set("fs.azure.account.oauth2.client.secret.adlfdevbankdomain.dfs.core.windows.net", "valueid")
spark.conf.set("fs.azure.account.oauth2.client.endpoint.adlfdevbankdomain.dfs.core.windows.net", "https://login.microsoftonline.com/tenanatid/oauth2/token")
```

### Read the Data

```python
df_users_data = spark.read.format("parquet").option("header", "true").option("inferSchema", "true").load('abfss://input@adlfdevbankdomain.dfs.core.windows.net/users_data.parquet')
```

```python
df_users_data.display()
```

### Identify the premium customers

```python
df_credit_score = df_users_data.filter(col("credit_score") >800).select("id", "credit_score").orderBy(col("credit_score"), ascending=False).display()
```
